```python
# import matplotlib

import matplotlib.pyplot as plt
```

```python
# Line plot

x = [1,2,3,4,5]
y = [10,12,8,14,10]

plt.plot(x,y,color='y',linestyle='-.',marker='o',linewidth=2)
plt.title("Line Plot")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

```python
# Bar chart

x = ['A','B','C','D']
y = [10,15,7,12]

plt.bar(x,y,color='skyblue')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid()
plt.show()
```

```python
# Scatter Plot

x = [1,2,3,4,5]
y = [10,12,8,14,10]

plt.scatter(x,y,color='r',marker='*')
plt.title("LScatter Plot")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid()
plt.show()
```
