```python
# import modules

import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Data

data = {
    'Customer ID':[1,2,3,4,5],
    'Age':[25,40,28,45,22],
    'Income':[50000,70000,52000,80000,45000],
    'SpendingScore':[65,40,60,30,80]
}

df = pd.DataFrame(data)
```

```python
# Cluster

x = df[['Age','Income','SpendingScore']]

kmeans = KMeans(n_clusters=3)
kmeans.fit(x)

df['Cluster'] = kmeans.labels_

print(df)
```

```python
# Plot

plt.figure(figsize=(8,6))
plt.scatter(df['Age'],df['SpendingScore'],c=df['Cluster'],cmap='viridis',s=100, marker='*')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Customer Clusters')
plt.grid()
plt.show()
```
